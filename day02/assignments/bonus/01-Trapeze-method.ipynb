```python
import numpy as np
import matplotlib.pyplot as plt

def compute_integrale_trapeze(x, y, nbi):
    
    integrale = 0.
    for i in range(nbi):
        integrale = integrale + (y[i]+y[i+1])*(x[i+1]-x[i])/2
   
    return integrale

def plot_integrale(x, y, nbi):
    
    for i in range(nbi):
        x_trapez = [x[i], x[i], x[i+1], x[i+1], x[i]] # Abscisses des sommets
        y_trapez = [0, y[i], y[i+1], 0, 0] # Ordonn√©es des sommets
        
        plt.plot(x_trapez, y_trapez, "b")
   
    #return 0
```

```python
xmin = 0
xmax = 3*np.pi/2
nbx = 100
nbi = nbx - 1 # nombre d'intervalles
```

- Consider function cosinus
- Write the function compute_integrale_trapeze to compute the integral as described on the lesson
- Write the function plot_integrale to plot the integral
- Plot both cosinus function and its integral

```python
%%time

x = np.linspace(xmin, xmax, nbx)
y = np.cos(x)
plt.plot(x,y,"ro-")
integral = compute_integrale_trapeze(x, y, nbi)
plot_integrale(x, y, nbi)
print("integrale =", integral)
plt.show()
```
